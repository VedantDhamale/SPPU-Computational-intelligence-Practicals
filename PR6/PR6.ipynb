```python
import numpy as np, matplotlib.pyplot as plt

def best_fitness_value(x): return sum(x**2)

def clonal_selection(pop, dim, gens):
    p = np.random.uniform(-5, 5, (pop, dim))
    hist = []
    for _ in range(gens):
        fit = np.array([best_fitness_value(i) for i in p])
        best = p[np.argmin(fit)]
        clone = best + np.random.uniform(-0.5, 0.5, dim)
        p[np.argmax(fit)] = clone
        hist.append(best_fitness_value(best))
    return best, hist


pop_size = 10
dim = 2
generations = 20

best, hist = clonal_selection(pop_size, dim, generations)
print("Best solution:", best)

plt.plot(hist, marker='o')
plt.title("Fitness over Generations")
plt.xlabel("Generation"); plt.ylabel("Fitness")
plt.grid(); plt.show()
#Clonal Selection
```
